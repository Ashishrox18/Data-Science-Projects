```python
import numpy as np
```

```python
arr = np.arange(1, 21)  # Array from 1 to 20
print("Array:", arr)

# a. Mathematical operations
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))

# b. Indices of elements > 10
indices = np.where(arr > 10)
print("Indices of elements greater than 10:", indices)
```

```python
arr_2d = np.arange(1, 17).reshape(4, 4)
print("4x4 Array:\n", arr_2d)

# a. Transpose
print("Transpose:\n", arr_2d.T)

# b. Row-wise and Column-wise sums
print("Row-wise sums:", np.sum(arr_2d, axis=1))
print("Column-wise sums:", np.sum(arr_2d, axis=0))
```

```python
A = np.random.randint(1, 21, (3, 3))
B = np.random.randint(1, 21, (3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# a. Element-wise operations
print("Addition:\n", A + B)
print("Subtraction:\n", A - B)
print("Multiplication:\n", A * B)

# b. Dot Product
print("Dot Product:\n", np.dot(A, B))
```

```python
arr12 = np.arange(1, 13)
reshaped = arr12.reshape(3, 4)
print("Reshaped Array:\n", reshaped)

# Slice: first two rows and last two columns
print("Sliced Array:\n", reshaped[:2, 2:])
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

print(df.head())  # a. First 5 rows

# b. Summary statistics
print(df[['Age', 'Salary']].describe())

# c. Average HR salary
avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average HR Salary:", avg_hr_salary)
```

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print(filtered_df)
```

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
sorted_df.to_csv("sorted_employee_data.csv", index=False)

print("CSV file saved successfully!")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks)
plt.title("Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1, 0, 0, 0]  # Highlight highest revenue region

plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title("Revenue Distribution by Region")
plt.show()
```

```python
data = np.random.randint(1, 101, 1000)

plt.hist(data, bins=10)
plt.title("Frequency Distribution of Random Numbers")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

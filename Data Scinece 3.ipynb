```python
import random

# Part a: Tossing a coin 10,000 times
trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

print("Coin Toss Simulation:")
print("Heads Probability:", heads / trials)
print("Tails Probability:", tails / trials)

# Part b: Rolling two dice and probability of sum = 7
trials_7 = 10000
count_sum_7 = 0

for _ in range(trials_7):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        count_sum_7 += 1

print("\nDice Roll Simulation:")
print("Probability of Sum = 7:", count_sum_7 / trials_7)
```

```python
import random

def prob_at_least_one_six(trials=10000):
    success = 0

    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1

    return success / trials

print("\nProbability of getting at least one 6 in 10 rolls:", prob_at_least_one_six())
```

```python
import random

# 5 red, 7 green, 8 blue → 20 total balls
colors = ['R']*5 + ['G']*7 + ['B']*8
trials = 1000

prev_blue = 0
red_given_blue = 0

results = []  # store (previous, current)

for _ in range(trials):
    prev = random.choice(colors)
    curr = random.choice(colors)
    results.append((prev, curr))

for prev, curr in results:
    if prev == 'B':
        prev_blue += 1
        if curr == 'R':
            red_given_blue += 1

P_R_given_B = red_given_blue / prev_blue

print("\nConditional Probability:")
print("P(Red | Blue) =", P_R_given_B)

# Compute overall simulation probabilities
P_R = sum(1 for r in results if r[1] == 'R') / trials
P_B = sum(1 for r in results if r[0] == 'B') / trials

# Bayes Theorem: P(R|B) ≈ P(B|R)*P(R) / P(B)
P_B_given_R = sum(1 for r in results if r[0] == 'B' and r[1] == 'R') / sum(1 for r in results if r[1] == 'R')
P_R_using_bayes = (P_B_given_R * P_R) / P_B

print("Bayes Theorem Check ≈", P_R_using_bayes)
```

```python
import numpy as np

values = [1, 2, 3]
prob = [0.25, 0.35, 0.40]

sample = np.random.choice(values, size=1000, p=prob)

print("\nDiscrete RV Results:")
print("Mean:", np.mean(sample))
print("Variance:", np.var(sample))
print("Standard Deviation:", np.std(sample))
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=5, size=2000)

plt.hist(data, bins=30, density=True, alpha=0.6, label="Histogram")
plt.title("Exponential Distribution (λ = 1/5)")
plt.xlabel("Value")
plt.ylabel("Density")

# PDF Overlay
x_values = np.linspace(0, max(data), 200)
lambda_param = 1/5
pdf = lambda_param * np.exp(-lambda_param * x_values)
plt.plot(x_values, pdf, label="PDF")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 random values
data = np.random.uniform(0, 1, 10000)

# Draw 1000 samples of size n=30 and store means
sample_means = []

for _ in range(1000):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))

# Plot original uniform distribution
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(data, bins=30, alpha=0.7)
plt.title("Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Plot distribution of sample means
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, alpha=0.7, density=True)
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Means")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```
